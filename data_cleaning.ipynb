```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df
```

```python
# Check which columns have missing values.
df.isnull().sum()
```

```python
# Remove Duplicates Values 
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
# Fill in The Values in Calories column by mean()
df["Calories"] = df["Calories"].fillna (df["Calories"].mean())
df['Calories'] = df['Calories'].round(2)
```

```python
df.head()
```

```python
df.info()
```

```python
# Fill in the Null values
df['Date'] = df['Date'].fillna('2012/10/04')

# Replace index wise Date Column
df.loc[26, "Date"] = '2020/12/26'
```

```python
df
```

```python
# Select records with pulse > 110 and Maxpulse < 150
data = df[(df['Pulse'] > 110) & (df['Maxpulse'] < 150)]
```

```python
data
```

```python
data["Calories"].mean()
```

```python
# Convert Colories coloum to descending order
df.sort_values(by='Calories',ascending=False).head()
```

```python
# Conver Date as string and remove Extra space and single quotes
df['Date'] = df['Date'].astype(str).str.strip().str.replace("'", "").str.replace('"', '')
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df
```

```python
# Reset index the index and staart numbering form 1
df.reset_index(drop=True, inplace=True)
df.index = df.index + 1
```

```python
# Add a new columns showing the day of the week from  the date column
df['Day of Week'] = df.Date.dt.day_name()
df.head()
```
